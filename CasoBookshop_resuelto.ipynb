```python
#Renombramos las columnas
df_store.rename(columns={
    'Title': 'Título',
    'Units sold': 'Cantidad vendida',
    'List price': 'Precio de venta',
    'Royalty paid': 'Royalty'
}, inplace=True)

df_store.head(10)
```

```python
#Eliminamos filas que tienen todos Nan
df_store.dropna(how='all', inplace=True)
df_store
```

```python
#Creamos un df con las filas de US
df_storeUS = df_store.iloc[0:8]
df_storeUS.reset_index(drop=True, inplace=True)
df_storeUS
```

```python
#Eliminamos las filas que no corresponden a títulos de los libros
df_storeUS = df_storeUS.drop([0,1])
df_storeUS.reset_index(drop=True, inplace=True)
df_storeUS
```

```python
#Creamos columna Store
df_storeUS['Store'] = 'US'
df_storeUS
```

```python
#Seleccionamos las filas de EUR
df_storeEUR = df_store.iloc[8::]
df_storeEUR
```

```python
#Eliminamos filas que no son titulos
df_storeEUR = df_storeEUR.drop([11,12])
df_storeEUR.reset_index(drop=True, inplace=True)
df_storeEUR
```

```python
#Eliminamos las que no son títulos
df_storeEUR = df_storeEUR.drop([4,5])
df_storeEUR
```

```python
#Creamos la columna
df_storeEUR['Store'] = 'EUR'
df_storeEUR
```

```python
#Concatenamos los 2 df limpios
df_Storenew = pd.concat([df_storeUS, df_storeEUR], ignore_index=True)
df_Storenew
```

```python
#creamos en el df_web la columna Store
df_web['Store'] = 'Web'
df_web
```

```python
#Cocatenamos los df store y web
df_final = pd.concat([df_Storenew, df_web], ignore_index=True)
df_final
```

```python
#Elimimnamos los Nan que no corresponden
df_final = df_final.dropna(subset=['Cantidad vendida'])
df_final
```

```python
#Reordenamos las columnas para que queden en el mismo orden que se solicita
df_final = df_final[['Store', 'Título', 'Cantidad vendida', 'Precio de venta', 'Royalty']]
df_final
```

```python
#Ordenamos la posición de la columna Store alfábeticamente
df_final = df_final.sort_values(by='Store', ascending=False)
df_final
```

```python
#Cambiamos el nombre de las columnas para que coincidan con las que se solicitan
df_final.rename(columns={
    'Store': 'Store',
    'Título': 'Title',
    'Cantidad vendida': 'Units sold',
    'Precio de venta': 'List price',
    'Royalty': 'Royalty'}, inplace=True)

df_final

```

```python
#Exportamos a excel
df_final.to_excel('bookshop-sales-data.xlsx', index=False)
```

```python
#Cuáles son los libros más vendidos?
más_vendidos = df_final.groupby('Title', as_index=False)['Units sold'].sum()
más_vendidos = más_vendidos.sort_values(by='Units sold', ascending=False)
más_vendidos
print(más_vendidos)
print(f'El libro más vendido es {más_vendidos.iloc[0,0]}')
```

```python
#Cuáles son los libros que más conviene vender
df_final['Total Sales'] = df_final['Units sold'] * df_final['List price']
mas_convenientes = df_final.groupby('Title', as_index=False)['Total Sales'].sum()
mas_convenientes = mas_convenientes.sort_values(by='Total Sales', ascending=False)
mas_convenientes
print(mas_convenientes)
```
