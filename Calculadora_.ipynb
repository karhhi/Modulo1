```python
import math

# ---- Calculadora normal ----
def calculadora_normal(a, b, operacion):
    if operacion == 1:      # suma
        return a + b
    elif operacion == 2:    # resta
        return a - b
    elif operacion == 3:    # multiplicación
        return a * b
    elif operacion == 4:    # división
        if b == 0:
            return "Error: división por cero"
        return a / b
    else:
        return "Operación no válida"


# ---- Calculadora científica ----
def calculadora_cientifica(a, b, operacion):
    if operacion == 1:      # potencia
        return a ** b
    elif operacion == 2:    # raíz cuadrada
        if a < 0:
            return "Error: no existe raíz real de número negativo"
        return math.sqrt(a)
    elif operacion == 3:    # logaritmo natural
        if a <= 0:
            return "Error: el logaritmo solo está definido para números positivos"
        return math.log(a)
    elif operacion == 4:    # factorial
        if a < 0 or int(a) != a:
            return "Error: factorial solo para enteros no negativos"
        return math.factorial(int(a))
    else:
        return "Operación no válida"


# ---- Programa principal ----
def main():
    tipo_calculadora = int(input("Ingrese el tipo de calculadora: 1: Normal  2: Científica: "))

    if tipo_calculadora == 1:
        # Calculadora normal
        a = float(input("Ingrese operando 1: "))
        b = float(input("Ingrese operando 2: "))
        operacion = int(input("Ingrese operación 1 = (+), 2 = (-), 3 = (*), 4 = (/): "))
        resultado = calculadora_normal(a, b, operacion)

    elif tipo_calculadora == 2:
        # Calculadora científica
        print("Operaciones científicas:")
        print("1 = Potencia (a^b)")
        print("2 = Raíz cuadrada de a")
        print("3 = Logaritmo natural de a")
        print("4 = Factorial de a")

        operacion = int(input("Seleccione operación: "))
        # Algunas usan 2 operandos, otras 1
        if operacion == 1:
            a = float(input("Ingrese base (a): "))
            b = float(input("Ingrese exponente (b): "))
        else:
            a = float(input("Ingrese número (a): "))
            b = None

        resultado = calculadora_cientifica(a, b, operacion)
    else:
        resultado = "Tipo de calculadora no válido"

    print("Resultado:", resultado)


# Ejecutar
main()

    
```
